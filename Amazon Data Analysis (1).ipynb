## Step 2: Data Cleaning
#### 2.1 Handle Missing Values

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Checking the duplicated rows 
df.duplicated().sum()
```

```python
# Fill missing rating_count with 0
df['rating_count'] = df['rating_count'].fillna('0')
```

```python
df.isna().sum()
```

### 2.2 Convert Data Types


```python
# Remove ₹ and convert prices to float
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Remove % from discount_percentage and convert to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
# Convert rating and rating_count to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(int)
```

### 2.3 Remove Duplicates

```python
# df.drop_duplicates(subset=['review_id'], keep='first', inplace=True)
```

```python
# ------------------------------
# 📌 4. Feature Engineering
# ------------------------------
# Calculate discount amount
df['Selling_amount'] = df['actual_price'] - df['discounted_price']

```

```python
df.head()
```

```python
df['category'].value_counts().head(30)
```

```python
# Extract main category from category column
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])
```

```python
# Display cleaned data
df.head()
```

```python
# Save cleaned dataset
df.to_csv("amazon_reviews_cleaned.csv", index=False)
```
