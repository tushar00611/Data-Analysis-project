#### ANALYSIS 1: Total Revenue

```python
order_items["final_price"] = (
    order_items["quantity"] * order_items["price"]
) - order_items["discount"]

total_revenue = order_items["final_price"].sum()
print("Total Revenue:", total_revenue)
```

#### ANALYSIS 2: Revenue by Category

```python
merged = (
    order_items
    .merge(products, on="product_id")
    .merge(categories, on="category_id")
)

category_revenue = (
    merged.groupby("name_y")["final_price"]
    .sum()
    .sort_values(ascending=False)
)

category_revenue
```

#### 📈 Visualization: Revenue by Category

```python
import matplotlib.pyplot as plt

category_revenue.plot(kind="bar", figsize=(10,5))
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

#### ANALYSIS 3: Monthly Sales Trend

```python
orders["order_date"] = pd.to_datetime(orders["order_date"])

sales_trend = (
    orders
    .merge(order_items, on="order_id")
    .groupby(pd.Grouper(key="order_date", freq="M"))["final_price"]
    .sum()
)

sales_trend
```

```python
sales_trend.plot(figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### ANALYSIS 4: Top 10 Customers by Spend

```python
customer_spend = (
    orders
    .merge(order_items, on="order_id")
    .groupby("customer_id")["final_price"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

customer_spend
```

### ANALYSIS 5: Product Ratings Distribution

```python
rating_dist = reviews["rating"].value_counts().sort_index()

rating_dist.plot(kind="bar")
plt.title("Product Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

#### ANALYSIS 6: Order Status Distribution

```python
order_status = orders["status"].value_counts()

order_status.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))
plt.title("Order Status Distribution")
plt.show()
```
